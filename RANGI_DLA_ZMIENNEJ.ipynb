## Rangi dla zmiennej

```python
# ******************************************************************************************************
# Obliczanie rang dla wartości zmiennej podanej w formie listy
# ******************************************************************************************************

# WPROWADZANIE DANYCH                              
# zmienna reprezentowana przez wartości na liście
# --------------------------                       
#                                                  <--- KROK 1  (wpisać dane do listy)
x_dane = [9,3,4,3,4,1,1,4,4,11,22,22,22,22,2,7,3]        # dane przykładowe

# x_dane = ['x','x','l','l','x','d','a','a','a','b','b']   # dane nominalne   
# x_dane = [5.1, 4.8, 4.7, 4.6, 5.2]                       # dane ilościowe 
# x_dane = [9,3,4,3,4,1,1,4,4,11,22,22,22,22,2,7,3]        # dane ilościowe lub nominalne
# --------------------------

# WYBÓR TYPU DANYCH                                
# n - dane NOMINALNE, KATEGORYCZNE
# q - dane ILOŚCIOWE, (WARTOŚCIOWE ?)
# --------------------------
typ_danych = 'n'           #                        <--- KROK 2  (wybrać typ danych)
# --------------------------
# ******************************************************************************************************
#                                     PROCESSING
x = []
for i, xi in enumerate(x_dane):          # utworzenie kopii danych na liście x
    x.append(xi)

idx = 0    
sum_i = 0
l_i = 0
R = []             # lista z obliczonymi ragami dla zmiennej liczbowej (ranga liczona wg nr wierszy)
Rzn = []           # lista z rangami dla zmiennej nominalnej, kategorycznej (liczona wg częstości zbiorów)
X = []
x.sort()
for i, xi in enumerate(x):
    l_i += 1
    sum_i += i+1
    if i < len(x)-1:           
        if x[i] != x[i+1]:
            R.append(sum_i/l_i)
            Rzn.append((l_i+1)/2)
            X.append(xi)
            sum_i = 0
            l_i = 0
    else:
        R.append(sum_i/l_i)
        Rzn.append((l_i+1)/2)
        X.append(xi)        

if typ_danych in['n', 'N']:
    print('Dane poniżej zostały potraktowane jako NOMINALNE/KATEGORYCZNE.')
if typ_danych in['q', 'Q']:
    print('Dane poniżej zostały potraktowane jako ILOŚCIOWE/WARTOŚCIOWE.')
print('#'*80, '\n')

print('dane wejściowe        =', x_dane)
print('dane uporządkowane    =', x)
print('Lista klastrów danych =', X)
if typ_danych in['q', 'Q']:
    print('Rangi ilościowe   =', R)
    print('\nZMIENNE ILOŚCIOWE/WARTOŚCIOWE\nUporządkowana lista rang dla zmiennej reprezentowanej przez podane wartości xi.')
    for i in range(len(X)):
        print(f'x{i+1} =', X[i], '    ranga =', R[i])

    print('\nRANGI dla ZMIENNYCH ILOŚCIOWYCH/WARTOŚCIOWYCH - kolejność podana w zadaniu (zmienna x_dane).')
    for i, xi in enumerate(x_dane):
        idx = X.index(xi)
    #     print(f'i={i}  x{i+1} =', x_dane[i], '      ranga =', R[idx])
        print(f'i={i}  x{i+1} =', xi, '      ranga =', R[idx])

if typ_danych in['n', 'N']:
    print('Rangi dla klastrów    =', Rzn)

    print('\nUporządkowana lista rang dla zmiennej nominalnej (kategorycznej) reprezentowanej przez podane wartości xi.')
    for i in range(len(Rzn)):
        print(f'x{i+1} =', X[i], '    ranga =', Rzn[i])    

    print('\nRANGI dla ZMIENNYCH NOMINALNYCH (kategorycznych) - kolejność podana w zadaniu (zmienna x_dane).')
    for i, xi in enumerate(x_dane):
        idx = X.index(xi)
        print(f'x{i+1} =', x_dane[i], '      ranga =', Rzn[idx])
print('\nListę rang należy przenieść do kratek odpowiadających wartościom xi w zadaniu. (Z.G.)')
    
```
